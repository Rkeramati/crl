```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
to_int = {'#': 0, ' ':1, 's': 2, 'g': 3, '0': 4, 'x': 5, 'd': 6, 'u': 7}
to_symbol = {v: k for k, v in to_int.items()}
def mv(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
n_trial = 5
res_do = []
for lamb in [10.0, 100.0, 1000.0]:
    add = 'results/method_DO/ent_known_1/random_0/map_name_maps/map_fourroom_exp1.txt/n_trial_5/n_sample_1/max_ep_1000/max_step_500/beta_10.0'
    add += '/lambd_{}'.format(lamb)
    count = np.load(add+'/results_count_trail_0_sample_0.npy')
    mapfile = np.load(add + '/results_map_sample_{}.npy'.format(0))
    annon = mapfile.reshape(11,11)[1:-1,1:-1]
    
    for n in range(1, n_trial):
        count += np.load(add+'/results_count_trail_{}_sample_0.npy'.format(n))
    
    res_do.append(add+'/results_results.npy')
    plt.figure(figsize=(5,5))
    plt.imshow((np.sum(count/n_trial, axis = 1)/np.sum(count)).reshape(11,11)[1:-1,1:-1])
    plt.colorbar()
    for i in range(9):
        
        for j in range(9):
            text = plt.text(j, i, to_symbol[annon[i, j]],
                           ha="center", va="center", color="w")
            plt.title('method DO, lambda = {}'.format(lamb))
```
