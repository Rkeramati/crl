```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mv(a, n=5) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
plt.figure(figsize=(15,10))
NoBonus1 = np.expand_dims(np.load('result/value_nS8NoBonus_0.npy'), axis=1)
NoBonus2 = np.expand_dims(np.load('result/value_nS8NoBonus_1.npy'), axis=1)
NoBonus3 = np.expand_dims(np.load('result/value_nS8NoBonus_2.npy'), axis=1)
for lambd in [0.1, 0.2, 0.7, 0.8]:
    Bonus1 = np.expand_dims(np.load('result/value_nS8Bonus_%d_0.npy'%(int(lambd*10))), axis=1)
    Bonus2 = np.expand_dims(np.load('result/value_nS8Bonus_%d_1.npy'%(int(lambd*10))), axis=1)
    Bonus3 = np.expand_dims(np.load('result/value_nS8Bonus_%d_2.npy'%(int(lambd*10))), axis=1)
    Bonus = np.concatenate([Bonus1, Bonus2, Bonus3], axis=1)
    plt.plot(mv(np.mean(Bonus, axis=1), 20), label='Bonus lambda %g'%(lambd))
for lambd in [0.02]:
    Bonus1 = np.expand_dims(np.load('result/value_nS8Bonus_0%d_0.npy'%(int(lambd*100))), axis=1)
    Bonus2 = np.expand_dims(np.load('result/value_nS8Bonus_0%d_1.npy'%(int(lambd*100))), axis=1)
    Bonus3 = np.expand_dims(np.load('result/value_nS8Bonus_0%d_2.npy'%(int(lambd*100))), axis=1)
    Bonus = np.concatenate([Bonus1, Bonus2, Bonus3], axis=1)
    plt.plot(mv(np.mean(Bonus, axis=1), 20), label='Bonus lambda %g'%(lambd))
NoBonus = np.concatenate([NoBonus1, NoBonus2, NoBonus3], axis=1)
plt.plot(mv(np.mean(NoBonus, axis=1), 20), label='NoBonus', linewidth=3)
plt.xlim([0, 500])
plt.legend(fontsize=20)
plt.xlabel('Episodes', fontsize=25)
plt.ylabel('Value', fontsize=25)
```
