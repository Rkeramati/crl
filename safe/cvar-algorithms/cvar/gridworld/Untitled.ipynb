```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
max_num = 500
value = np.zeros((max_num, 50))
for it in range(max_num):
    value[it, :] = np.load('cvars_%d.npy'%(it))
```

```python
value.shape
```

```python
some = [0, 10, 40]
plt.figure(figsize = (15, 10))
for c in some:
    plt.plot(value[:, c], label = 'alpha = %g'%(c*0.01), linewidth = 3)
plt.legend(fontsize = 15)
plt.xlabel('epsiodes', fontsize = 15)
plt.ylabel('CVaR x(0) at alpha 0.5', fontsize = 15)
```

```python
plt.plot(np.load('returns.npy'))
```

```python
max_num = 1000
value = np.zeros((max_num, 50))
for it in range(max_num):
    value[it, :] = np.load('cvars_small_%d.npy'%(it))
```

```python
some = [0, 10]
plt.figure(figsize = (15, 10))
for c in some:
    plt.plot(value[:, c], label = 'alpha = %g'%(c*0.01), linewidth = 3)
plt.plot(np.ones(1000) * -38.05, label = 'True CVaR')
plt.legend(fontsize = 15)
plt.plot()
plt.xlabel('epsiodes', fontsize = 15)
plt.ylabel('CVaR x(0) at alpha 0.5', fontsize = 15)
```
