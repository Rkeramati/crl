```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
plt.plot(moving_average(np.max(cvar_opt, axis=0), 100), label='opt')
plt.plot(moving_average(np.max(cvar, axis=0), 100), label='e-greedy')
plt.legend()
```
